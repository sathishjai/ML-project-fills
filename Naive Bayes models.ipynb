```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data=pd.read_csv('/home/sathish/Downloads/app_review.csv')
data
```

```python
def preprocess_data(data):

    data=data.drop('package_name',axis=1)

    data['review']=data['review'].str.strip().str.lower()
    return data
data=preprocess_data(data)
data.head()
```

```python
x=data['review']
y=data['polarity']
x,x_test,y,y_test=train_test_split(x,y,stratify=y,test_size=0.25,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Assuming x is a list of strings
vec = CountVectorizer(stop_words='english')
x = vec.fit_transform(x).toarray()
x_test = vec.transform(x_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model1=MultinomialNB()
model1.fit(x,y)
```

```python
print(model1.score(x_test,y_test))
print(model1.predict(vec.transform(['Love this app simply awessome!'])))
print(model1.predict(vec.transform(['This word is used to describe a review that is strongly critical or unfavorable.'])))
```
