## find the prime number in 1 to 100 and condition for do not include 1,3 and 7 another number show output

```python
for i in range(1,101):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        val=str(i)
        if '1' in val or '3' in val or '7' in val:
            pass
        else:
            print(val)
```

## Naive Bayes models

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data=pd.read_csv('/home/sathish/Downloads/archive(72)/Instagram.csv')
data
```

```python
data.isnull().sum()
```

```python
# drop rows with missing values

data = data.dropna()


# create a CountVectorizer instance

#vec = CountVectorizer(stop_words='english')


# fit and transform the data

#data = vec.fit_transform(data['text']).toarray()
```

```python
def preprocess_data(data):

    data=data.drop('reviewId',axis=1)

    data['content']=data['content'].str.strip().str.lower()
    return data
data=preprocess_data(data)
data.head()
```

```python
data.isnull().sum()
```

```python
x=data['content']
y=data['score']
x,x_test,y,y_test=train_test_split(x,y,stratify=y,test_size=0.25,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Assuming x is a list of strings
vec = CountVectorizer(stop_words='english')
x = vec.fit_transform(x).toarray()
x_test = vec.transform(x_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model1=MultinomialNB()
model1.fit(x,y)
```

```python
print(model1.score(x_test,y_test))
print(model1.predict(vec.transform(['Love this app simply awessome!'])))
print(model1.predict(vec.transform(['This word is used to describe a review that is strongly critical or unfavorable.'])))
```
