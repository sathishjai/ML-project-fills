```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/home/sathish/Downloads/swedish_insurance_univariate.csv')
```

```python
dataset
```

```python
x=dataset[['X']]
y=dataset['Y']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
k=2
knn=KNeighborsRegressor(n_neighbors=k)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(f"Mean_sqared error(mse) : {mse:.2f}")
print(f"R_sqared(R2) score: {r2: .2f}")
```

```python
plt.scatter(x_test,y_test,label="Actual")
plt.scatter(x_test,y_pred,label="predicted",color='red')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"KNN Regression (k={k})")
plt.show()
```
